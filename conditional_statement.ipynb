conditional statement:

```python
## If statement:
age = int(input("Enter your age: "))
if age>=20:
    print("you are allowed to vote")
```

Else statement:
The else statement executes the block of code if the condition in the if statement is false.

```python
age = int(input("Enter your age: "))
if age >= 18:
    print("you are eligible to vote")
else:
    print("you are not eligible to vote")
```

Elif:
1. Stands for elseif
2. Used to check multiple conditions

```python
age = int(input("Enter your age: "))
if age < 13:
    print("You are a child.")
elif age < 18:
    print("You are a teenager.")    
else:
    print("You are an adult.")
```

Nested conditional statement:
we can place one or more if, elif or else statement inside another if, elif or else statement to create a nested conditional statements.

```python
num=int(input("Enter a number: "))
if num>0:
    print("The number is positive.")
    if num%2==0:
        print("The number is even.")
    else:
        print("The number is odd.")
else:
    print("The number is zero or negative .")
```

```python
## Practical example:
## Determine if a year is a leap year or not.
year = int(input("Enter a year: "))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(f"{year} is a leap year.")
        else:
            print(f"{year} is not a leap year.")
    else:
        print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
## simple calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter the operation (+, -, *, /): ")

if operation == "+":
    result = num1 + num2
if operation == "-":
    result = num1 - num2
if operation == "*":
    result = num1 * num2
if operation == "/":
    if num2!=0:
        result = num1 / num2
    else:
        result = "Error! Division by zero."

else:
    result = "Invalid operation."

print("Result: ", result)
```

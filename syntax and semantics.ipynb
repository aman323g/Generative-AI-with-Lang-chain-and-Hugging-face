Data Types in python:

1.Data type are a classification of data which tells compiler or intereter how the programmer intend to use the data.
2.Also determine what tupe of operation performed on Data, the values that can be taken and the amount of memory needed to store the data.

Importance:

1.It ensures that the data stored in efficient way.
2.Helps in performing correct operations on data.
3.Proper use of data types can prevent errors and bugs in program.




```python
## Examples of Data types
age=35               ## Integer
height=5.9          ## Float    
name="Aman"         ## String
type(True)          ## Boolean
is_true=True      ## Boolean
print(is_true)
print(type(is_true))
```

```python
a=10
b=10
a==b
type(a==b)
print(type(a==b))
```

```python
result = "Hello" + " " +str(5)
print(result)
```

```python
## Arithmetic Operators 
a = 10
b = 5
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b) ## Actual Division
print("Floor Division:", a // b) ## Round off value
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
```

```python
## Comparison operators
a = 10
b = 10
result = (a == b)
print(result)
```

Logical oerator

```python
x = True
Y = True
result1 = (x and Y)
print(result1)
result2 = (x or Y)
print(result2)
result3 = not x
print(result3)
```
